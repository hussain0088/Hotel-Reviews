**Stop Words Removel**

```python
!pip install contractions
import contractions

# Load English stopwords and keep 'not'
en_stopwords = set(stopwords.words('english'))
en_stopwords.discard('not')
df['clean_review'] = df['Review'].apply(lambda review: ' '.join(
    [word for word in contractions.fix(review).split() if word not in en_stopwords]
))
```

```python
df.head()
```

**Remove Punctuation**

```python
df['clean_review_no_punc']=df.apply(lambda x: re.sub(r"[*]", 'star',x ['clean_review']), axis=1)
```

```python
df.head()
```

```python


df['clean_review_no_punc'] = df['clean_review_no_punc'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
```

```python
df.head()
```

**Tokenization**

```python

df['tokenize']=df.apply(lambda x: word_tokenize(x['clean_review_no_punc']), axis=1)
```

```python
df['tokenize'][0]
```

**Stemming**

```python
ps=PorterStemmer()
df["stemmed"]=df['tokenize'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
df.head()
```

**Lemmatization**

```python

lemmatizer=WordNetLemmatizer()
df["lemmatize"]=df['tokenize'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
df.head()
```

**N-grams**

```python
token_clean=sum(df['lemmatize'],[])
print(token_clean)
```

```python
unigram=(pd.Series(nltk.ngrams(token_clean, 1)).value_counts())
```

```python
print(unigram)
```

```python
bigram=(pd.Series(nltk.ngrams(token_clean, 2)).value_counts())
print(bigram)
```

```python
trigram=(pd.Series(nltk.ngrams(token_clean, 3)).value_counts())
print(trigram)
```
